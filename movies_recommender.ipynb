Import libraries and data

```python
import pandas as pd

links = pd.read_csv('ml-latest-small/links.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
ratings = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
#check for missing values
ratings.isnull().sum()
```

```python
user_ratings = pd.pivot_table(data=ratings, values='rating', index='userId', columns='movieId')
user_ratings
```

```python
#replace NaNs with 0
user_ratings.fillna(0, inplace=True)
user_ratings
```

Compute cosine similarity matrix

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarities = pd.DataFrame(cosine_similarity(user_ratings), columns = user_ratings.index, index=user_ratings.index)
user_similarities
```

Recommender systems by steps

```python
#compute the weights
user_id = 1

weights = (
    user_similarities.query('userId!=@user_id')[user_id] / sum(user_similarities.query('userId != @user_id')[user_id])
)
weights #weights for each user sums to 1. You can check by running weights.sum()
```

```python
#exclude our user, since we don't want to include them on the weights.
user_ratings.loc[user_id, :] == 0
```

```python
# select movies that user 1 has not rated

user_unrated_movies = user_ratings.loc[user_ratings.index != user_id, user_ratings.loc[user_id, :] == 0]

#By calling .T on this subset, the data frame is transposed so that the rows represent movies 
# and the columns represent users. This makes it easier to calculate the weighted average rating for each unrated movie, 
# as the dot product of the transposed subset and the weights vector will give a weighted average rating for each movie.
user_unrated_movies.T
```

```python
# Compute the ratings user 1 would give to those unrated movies.
# dot product between the unrated movies and the weights

weighted_averages = pd.DataFrame(user_unrated_movies.T.dot(weights), columns=['predicted_rating'])
weighted_averages
```

```python
# sort the predicted ratings in descending order
movie_recommendations = weighted_averages.merge(movies, left_index=True, right_on='movieId').sort_values('predicted_rating', ascending=False)
movie_recommendations
```

function for the movies recommender

```python

def recommend_movies(userId, n):
    # compute the weights
    weights = (
        user_similarities.query('userId != @userId')[userId] / sum(user_similarities.query('userId != @userId')[userId])
    )
    user_unrated_movies = user_ratings.loc[user_ratings.index != userId, user_ratings.loc[userId, :] == 0]
    weighted_averages = pd.DataFrame(user_unrated_movies.T.dot(weights), columns=['predicted_rating'])
    movie_recommendations = weighted_averages.merge(movies, left_index=True, right_on='movieId').sort_values('predicted_rating', ascending=False)
    return movie_recommendations.head(n)
```

```python
recommend_movies(24, 6)
```
